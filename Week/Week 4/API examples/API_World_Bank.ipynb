```python
import requests
```

```python
countries=['AUS','BRA']
```

```python
data=[]
for country in countries:
  url='https://api.worldbank.org/v2/country/'+country+'/indicator/SH.STA.ANVC.ZS?format=json'
  json_data=requests.get(url).json()[1]
  data.append(json_data)
```

```python
import numpy as np
```

```python
flat_data=np.array(data).flatten()
```

```python
df=pd.json_normalize(flat_data)
df
```
